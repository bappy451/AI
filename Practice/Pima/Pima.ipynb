```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.models import Model
from keras.optimizers import Adam,RMSprop
from sklearn.model_selection import StratifiedKFold
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
detaset = np.loadtxt('pima.csv', delimiter=',')
```

```python
X = detaset[:,0:8]
y = detaset[:,8]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3,shuffle=True)
```

```python
def create_model_m(optimizer='adam',init='uniform'):
    model = Sequential()
    model.add(Dense(16,input_dim=8, kernel_initializer='uniform',activation='relu'))
    model.add(Dense(16, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model_m, epochs=100, batch_size=10)
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True)
```

```python
result = cross_val_score(model, X,y, cv=kfold,verbose=2)
```
