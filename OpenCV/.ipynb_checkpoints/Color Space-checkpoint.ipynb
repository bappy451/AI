# Color Space

```python
import cv2
import numpy as np

image = cv2.imread('./images/input.jpg')
```

## Important Note
### In openCV images are BGR but in python and matplotlib images are RGB

```python
import matplotlib.pyplot as plt
```

```python
# as we are shoing througn plt this is not original view
plt.imshow(image)
```

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# this is original view of the image
plt.imshow(rgb_img)
```

```python
# converting to gray
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

### Another useful color space is HSV(Hue Saturation Value)
Infact HSV is very useful in color filtering.

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
cv2.imshow('HSV image', hsv_image)
cv2.waitKey()
cv2.imshow('Hue channel', hsv_image[:, :, 0])
cv2.waitKey()
cv2.imshow('Saturation channel', hsv_image[:, :, 1])
cv2.waitKey()
cv2.imshow('Value channel', hsv_image[:, :, 2])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.subplot(311)
plt.imshow(hsv_image[:, :, 0],cmap='gray')
plt.subplot(312)
plt.imshow(hsv_image[:, :, 1],cmap='gray')
plt.subplot(313)
plt.imshow(hsv_image[:, :, 2],cmap='gray')
plt.show()
```

```python
plt.imshow(hsv_image,cmap='gray')
```

# OpenCV's 'split' function splites the image into each color index

```python
B, G, R = cv2.split(image)
```

```python
print(B.shape)
cv2.imshow("Red", R)
cv2.imshow("Green", G)
cv2.imshow("Blue", B)
cv2.waitKey()
cv2.destroyAllWindows()

# Let's re-make the original image, 
merged = cv2.merge([B, G, R]) 
cv2.imshow("Merged", merged) 

# Let's amplify the blue color
merged = cv2.merge([B+100, G, R])
cv2.imshow("Merged with Blue Amplified", merged) 

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(B,cmap='gray')
```

```python
plt.imshow(G,cmap='gray')
```

```python
plt.imshow(R,cmap='gray')
```
