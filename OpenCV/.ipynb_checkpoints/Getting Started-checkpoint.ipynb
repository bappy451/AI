```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading and showing image from images folder
# As waitKey() have no argument this will wait for any key
input_image = cv2.imread('./images/input.jpg')

cv2.imshow('Hello World', input_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#showing image shape 
#this is done by taking image as numpy array and its builtin function shape
print(input_image.shape)
```

```python
# Dimension of the image 
# If there is no color channel or image is grayscle and 2 dimention this will show error

print('Height of Image: {} pixels'.format(input_image.shape[0]))
print('Width of Image: {} pixels'.format(input_image.shape[1]))
print('Color channl: {}'.format(input_image.shape[2]))
```

```python
# save image as output.png
cv2.imwrite('output.png', input_image)
```

# Grayscaling
Grayscalling is process by which image is converted from a full color to shaded gray(Black & White)
In openCV builtin function can easily transform a image into graysclae form.

```python
# this will frist load a image and show that
# then convert that image to gray and show that
# Load our input image
image = cv2.imread('./images/input.jpg')
cv2.imshow('Original', image)
cv2.waitKey()

# We use cvtColor, to convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale', gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Another method faster method
# Showing inline with matplotlib
img = cv2.imread('./images/input.jpg',0)

plt.imshow(img,cmap='gray')
```
